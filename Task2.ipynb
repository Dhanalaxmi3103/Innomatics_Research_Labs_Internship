```python
#Defanging an IP address
#Given a valid (IPv4) IP address, return a defanged version of that IP address.

#A defanged IP address replaces every period "." with "[.]".

class Solution:
    def defangIPaddr(self, address: str) -> str:
        return address.replace(".","[.]")
        
```

```python
obj = Solution()
print(obj.defangIPaddr("1.1.1.1"))
print(obj.defangIPaddr("255.100.50.0"))

```

```python
#Find Numbers with an Even Number of digits
#Given an array nums of integers, return how many of them contain an even number of digits.
class Solution:
    def findNumbers(self, nums: list[int]) -> int:
        count = 0
        for num in nums:
            if len(str(num)) % 2 == 0:
                count += 1
        return count

```

```python
obj=Solution()
print(obj.findNumbers([12,345,2,6,7896]))
print(obj.findNumbers([555,901,482,1771]))
```

```python
#Number of Good Pairs
#Given an array of integers nums, return the number of good pairs.

#A pair (i, j) is called good if nums[i] == nums[j] and i < j.
class Solution:
    def numIdenticalPairs(self, nums: list[int]) -> int:
        count=0
        for i in range(len(nums)):
            for j in range(i+1,len(nums)):
                if nums[i]==nums[j]:
                    count+=1
        return count
```

```python
obj=Solution()
print(obj.numIdenticalPairs([1,2,3,1,1,3]))
print(obj.numIdenticalPairs([1,1,1,1]))
print(obj.numIdenticalPairs([1,2,3]))
```

```python
#How many numbers are smaller than the current number
#Given the array nums, for each nums[i] find out how many numbers in the array are smaller than it. 
#That is, for each nums[i] you have to count the number of valid j's such that j != i and nums[j] < nums[i].

#Return the answer in an array.
class Solution:
    def smallerNumbersThanCurrent(self, nums: list[int]) -> list[int]:
        result=[]
        for i in range(len(nums)):
            count=0
            for j in range(len(nums)):
                if nums[j]<nums[i]:
                    count+=1
            result.append(count)
        return result
```

```python
obj=Solution()
print(obj.smallerNumbersThanCurrent([8,1,2,2,3]))
print(obj.smallerNumbersThanCurrent([6,5,4,8]))
```

```python
#Subract the Product and Sum of Digits of an Integer
#Given an integer number n, return the difference between the product of its digits and the sum of its digits.
class Solution:
    def subtractProductAndSum(self, n: int) -> int:
        product=1
        sum_total=0
        while n>0:
            digit=n%10
            product*=digit
            sum_total+=digit
            n=n//10
        return product-sum_total
```

```python
obj=Solution()
print(obj.subtractProductAndSum(245))
print(obj.subtractProductAndSum(541))
print(obj.subtractProductAndSum(4421))
```

```python
#XOR Operation in an Array
#You are given an integer n and an integer start.Define an array nums where nums[i] = start + 2 * i (0-indexed) and n == nums.length.
#Return the bitwise XOR of all elements of nums.
class Solution:
    def xorOperation(self, n: int, start: int) -> int:
        result=0
        for i in range(n):
            result ^=(start+2*i)
        return result
```

```python
obj=Solution()
print(obj.xorOperation(5,0))
print(obj.xorOperation(4,3))
```
