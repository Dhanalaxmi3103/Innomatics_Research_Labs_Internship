```python
#Running Sum of 1D arry
def runningSum(arr):
    result=[]
    total=0
    for i in arr:
        total+=i
        result.append(total)
    return result
```

```python
arr=[1,3,6,9,2]
runningSum(arr)
```

```python
arr1=[4,6,8,10,12,16]
runningSum(arr1)
```

```python
#Shuffle of the Array
def shuffleArray(array,n):
    result=[]
    for i in range(n):
        result.append(array[i])
        result.append(array[i+n])
    return result
```

```python
arr=[1,3,5,9,8,7]
shuffleArray(arr,3)
```

```python
arr1=[1,3,5,7,10,12,14,16]
shuffleArray(arr1,4)
```

```python
#Kids with the Greatest Number of Candies
def kidsWithGreatestCandies(candies, extraCandies):
    max_candies = max(candies)
    result = []
    
    for c in candies:
        result.append(c + extraCandies >= max_candies)
    
    return result
```

```python
candies = [2,3,5,1,3]
extraCandies = 3
kidsWithGreatestCandies(candies, extraCandies)
```

```python
candies=[4,6,8,2,3,1,6,9,5]
extraCandies=5
kidsWithGreatestCandies(candies, extraCandies)
```
